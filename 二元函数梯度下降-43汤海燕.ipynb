```python
import numpy as np
 
# 定义二元函数
def f(x, y):
    return (x - 2)**2 + (y - 3)**2 + 10 * ((x + 2)**2 + (y - 1)**2)
 
# 定义梯度函数
def gradient(x, y):
    df_dx = 2 * (x - 2) + 20 * (x + 2)
    df_dy = 2 * (y - 3) + 20 * (y - 1)
    return np.array([df_dx, df_dy])
 
# 梯度下降算法
def gradient_descent(initial_point, learning_rate, num_iterations):
    point = np.array(initial_point, dtype=float)
    for i in range(num_iterations):
        grad = gradient(point[0], point[1])
        point -= learning_rate * grad
        if i % 100 == 0:  # 每100次迭代打印一次状态
            print(f"Iteration {i}: x = {point[0]}, y = {point[1]}, f(x, y) = {f(point[0], point[1])}")
    return point
 
# 初始点、学习率和迭代次数
initial_point = np.array([0.0, 0.0])
learning_rate = 0.01
num_iterations = 10000
 
# 运行梯度下降算法
minimum_point = gradient_descent(initial_point, learning_rate, num_iterations)
print(f"Approximate global minimum point: ({minimum_point[0]}, {minimum_point[1]})")
print(f"Approximate global minimum value: {f(minimum_point[0], minimum_point[1])}")
```
